```python
#outlier removal method--IQR(inter quartile range)
```

```python
q1 = dataset["Estimated Income"].quantile(0.25)
q3 = dataset["Estimated Income"].quantile(0.75)
```

```python
#Function Transformer
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_excel(r"C:\Users\Lenovo\Downloads\banking Data03.xlsx")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
sns.distplot(dataset["Saving Accounts"])
plt.show()
```

```python
q1 = dataset["Saving Accounts"].quantile(0.25)
q3 = dataset["Saving Accounts"].quantile(0.75)
iqr = q3 - q1
```

```python
min_r = q1 - (1.5*iqr)
max_r = q3 + (1.5*iqr)
min_r,max_r
```

```python
dataset=dataset[dataset["Saving Accounts"]<=max_r]
```

```python
#to convert in normal distribution ,we use function transformer
from sklearn.preprocessing import FunctionTransformer
```

```python
ft = FunctionTransformer(func=np.log1p) #1p = p-1
ft.fit(dataset[["Saving Accounts"]])
```

```python
dataset["Saving Accounts_tf"]=ft.transform(dataset[["Saving Accounts"]])
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.distplot(dataset["Saving Accounts"])
plt.title("before")

plt.subplot(1,2,2)
sns.distplot(dataset["Saving Accounts_tf"])
plt.title("After")

plt.show()
```
